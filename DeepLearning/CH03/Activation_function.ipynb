```python
import numpy as np
import matplotlib.pylab as plt

''' step function '''
def step_function(x):
    return np.array(x > 0, dtype = np.int)

''' sigmond function '''
def sigmond(x):
    return (1 / (1 + np.exp(-x)))

''' ReLU function '''
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)

plt.title('Activative Function')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, step_function(x), color='blue', label='step function')
plt.plot(x, sigmond(x), color = 'red', label='sigmond function')
plt.plot(x, relu(x), color = 'lime', label='ReLU function')
plt.ylim(-0.1, 1.1)
plt.legend()

plt.show()
```
