```python
import sys, os
sys.path.append(os.pardir)
import numpy as np
from dataset.mnist import load_mnist
```

```python
def cross_entropy_error(y, t):
    if y.ndim == 1:
        t = t.reshape(1, t.size)
        y = y.reshape(1, y.size)
        
    batch_size = y.shape[0]
    return -np.sum(np.log(y[np.arange(batch_size), t])) / batch_size
```

```python
''' Load data '''
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, one_hot_label=True)
```

```python
''' image (28bit x 28bit = 784) '''
print ("shape of x_train: ", x_train.shape)
'''　Handwritten numerical recognition(0, 1, 2, ..... , 7, 8, 9) '''
print ("shape of t_train:", t_train.shape)

print ("shape of x_test: ", x_test.shape)
print ("shape of t_test:", t_test.shape)
```

```python
''' Extract 10 pieces randomly from the training data. '''
train_size = x_train.shape[0]
batch_size = 10

batch_mask = np.random.choice(train_size, batch_size)

print ("Output the extracted data = ", batch_mask)

for i in batch_mask:
    print ("x_train = ", x_train[i, :])
    a = np.amax(t_train[i], axis=None)
    
    counter = 0
    for j in t_train[i]:
        if a == j:
            print ("Recognized numbers　= ", counter)
        counter += 1
```
