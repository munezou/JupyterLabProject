```python
'''
----------------------------------------
import library
----------------------------------------
'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import confusion_matrix
```

```python
'''
-----------------------------------------
Parameter space to search
-----------------------------------------
'''
def param():
  ret = {
      'C':[1, 10, 100],
      'kernel':['rbf', 'linear', 'poly'],
      'degree':np.arange(1, 6, 1),
      'gamma':np.linspace(0.01, 1.0, 50)
  }
  return ret
```

```python
# prepare data
x = load_iris().data
y = load_iris().target
```

```python
# Get 25% test data of the whole
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
# Create an instance of GridSearchCV, learn and cord the score.
gscv = GridSearchCV(SVC(), param(), cv=4, verbose=2)
gscv.fit(x_train, y_train)
```

```python
# Get a list of scores.
gs_result = pd.DataFrame.from_dict(gscv.cv_results_)
gs_result.to_csv('gs_result.csv')
```

```python
# Obtain the highest performance model and classify the test data.
best = gscv.best_estimator_
pred = best.predict(x_test)
```

```python
plt.scatter(pred, y_test)
```

```python
# Output a confusion matrix.
print(confusion_matrix(y_test, pred))
```
